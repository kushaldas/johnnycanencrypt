```python
import johnnycanencrypt as jce
ks = jce.KeyStore("./store")
```

```python
k = ks.create_newkey("redhat", ["Live Coding <fossasia@kushaldas.in>"], jce.Cipher.Cv25519)
with open(f"{k.fingerprint}.pub", 'w') as f:
    f.write(k.get_pub_key())

print(k.fingerprint)
```

```python
with open(f"{k.fingerprint}.sec", 'w') as f:
    f.write(k.keyvalue.decode("utf-8"))
```
